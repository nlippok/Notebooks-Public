<h3>Description</h3>

<b>Interactive tools</b>

Depicted above is a oversimplified CR-OCT system made of a frequency comb source, interferometer and quadrature signal generation. The system can be controlled by modifying the FC source pulse width, optical delay as well as toggeling I/Q and background substruction. 

| Variable name | Description|
| :---: | :--- |
| PulseWidth | Changes the pulse width (FWHM). Pulse FWHM is normalized to the pulse period, $t_p$.|
| $2\tau\Delta\nu$ | Delay, $\tau$, normalized to the circular delay range, $(2\Delta\nu)^{-1}$. |
| I/Q | Quadrature signal on and off. |
| BG | Background subtraction on and off. |
| Log | Switches between log and linear scale. |

<b>Signal output</b>
<ul type="square">
    <li>The main graph shows the RF frequency space of a CR-OCT signal.</li>
    <li>The x-axis of the main graph represents the normalized RF frequency with respect to the optical Nyquist frequency, $F_{oN}$.</li>
    <li>The blue shaded area indicates the baseband, ranging $\pm(2t_p)^{-1}=\pm F_{oN}$ in the RF space, which corresponds to the circular delay range $\pm(2\Delta\nu)^{-1}$.</li>
    <li>The brown, dotted curve depicts the signal envelope, $G(f)$.</li>
    <li>The small inset graph shows a pulse shape function, $g(t)$.</li>
</ul>
